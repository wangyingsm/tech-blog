```python
src = ''
with open('day5.txt') as fn:
    src = fn.read().rstrip()
src
```

```python
def react(src):
    stable = False
    while not stable:
        stable = True
        i = 0
        while i<len(src)-1:
            if abs(ord(src[i])-ord(src[i+1])) == ord('a') - ord('A'):
                src = src[:i] + src[i+2:]
                stable = False
                continue
            i += 1
    return src
len(react(src))
```
