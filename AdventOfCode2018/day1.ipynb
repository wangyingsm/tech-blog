```python
alllines = ''
with open('day1.txt') as fn:
    alllines = list(map(lambda x: int(x), fn.readlines()))
alllines
```

```python
result = sum(alllines)
result
```

```python
freqs = set()
now = 0
found = False
while True:
    for f in alllines:
        now += f
        if now in freqs:
            found = True
            break
        freqs.add(now)
    if found:
        break
now
```
