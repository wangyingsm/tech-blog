```python
alllines = ''
with open('day2.txt') as fn:
    alllines = list(map(lambda x: x.rstrip(), fn.readlines()))
alllines
```

```python
twice, trice = 0, 0
for l in alllines:
    lc = {}
    hastwice, hastrice = False, False
    for c in l:
        lc.setdefault(c, 0)
        lc[c] += 1
    for k, v in lc.items():
        if v == 2: hastwice = True
        if v == 3: hastrice = True
    if hastwice: twice += 1
    if hastrice: trice += 1
trice*twice
```

```python
def strdiff(s1, s2):
    diff = 0
    common = ''
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            diff += 1
        else:
            common += s1[i]
    return diff, common
```

```python
mindiff = 2**10
common = ''
for i in range(len(alllines)):
    for j in range(i+1, len(alllines)):
        d, c = strdiff(alllines[i], alllines[j])
        if d < mindiff:
            mindiff, common = d, c
common
```
