```python
def input(filename='day9.txt'):
    with open(filename) as fn:
        line = fn.read().rstrip().split(' ')
    return int(line[0]), int(line[6])
```

```python
class Circle():
    def __init__(self):
        self.marbles = [0]
        self.current = 0
    def putMarble(self, n):
        if len(self.marbles) == 1:
            self.marbles.append(n)
            self.current = 1
            return 0
        if n % 23 == 0:
            pos = (self.current - 7) % len(self.marbles)
            score = n + self.marbles[pos]
            self.marbles = self.marbles[:pos] + self.marbles[pos+1:]
            self.current = pos
            return score
        pos = (self.current + 2) % len(self.marbles)
        if pos > 0:
            self.marbles.insert(pos, n)
            self.current = pos
        else:
            self.marbles.append(n)
            self.current = len(self.marbles) -1
        return 0
    def __repr__(self):
        def tostr(x):
            if x[0] == self.current:
                return f'({x[1]})'
            return f'{x[1]}'
        return ' '.join(map(tostr, enumerate(self.marbles)))
```

```python
c = Circle()
for i in range(1, 1024):
    c.putMarble(i)
    print(c)
```

```python
def playerRun(players, maxMarble):
    import numpy as np
    scores = np.zeros((players), dtype=np.int)
    c = Circle()
    for rounds in range(maxMarble-1):
        scores[rounds%players] += c.putMarble(rounds+1)
    return scores
print(playerRun(9, 25).max())
print(playerRun(10, 1618).max())
print(playerRun(13, 7999).max())
print(playerRun(17, 1104).max())
print(playerRun(21, 6111).max())
print(playerRun(30, 5807).max())
```
