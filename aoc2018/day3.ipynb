```python
import numpy as np
```

```python
class Square():
    def __init__(self, sid, left, top, width, height):
        self.sid = sid
        self.left = left
        self.top = top
        self.width = width
        self.height = height
```

```python
def init_frabrics():
    frabrics = np.zeros((1000, 1000), dtype=np.int)
    return frabrics
init_frabrics()
```

```python
def fill_square(square, frabrics):
    l, t = square.left-1, square.top-1
    w, h = square.width, square.height
    frabrics[t:t+h, l:l+w] += 1
    return frabrics
```

```python
s = Square(1, 1, 2, 2, 5)
f = init_frabrics()
fill_square(s, f)
```
