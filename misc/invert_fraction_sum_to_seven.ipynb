```python
from fractions import Fraction as fr
```

```python
# 先不考虑必须大于100的情况
a = fr(1, 7)
while a>1e-10:
    d = int(a.denominator/a.numerator)+1
    b = fr(1, d)
    a -= b
    print(b)
    if a.numerator==1 and a.denominator>100:
        print(a)
        break
```

```python
# 验证结果
fr(1,8)+fr(1,57)+fr(1,3192)
```

```python
result = [] # 全局变量用来记录所有的结果分数
```

```python
def find_deno(x):
    '''编写一个函数求所有大于100的倒数分数'''
    global result
    while True:
        d = int(1/x)+1 #从下一个最接近的倒数分数开始减除
        while fr(1, d) in result: d+=1 # 如果该分数已经存在结果中，加一
        b = fr(1, d)
        x -= b # 减除该分数
        if d<=100: # 如果该分数分母小于等于100，递归求值
            find_deno(b)
        else:
            result.append(b) # 将该减除的分数加入结果中
            if x.denominator>100 and x.numerator==1 and x not in result:
                # 如果减去剩余的分数已经满足分子为1分母大于100，且未存在于结果中，将剩余分数加入结果中
                result.append(x)
                break
```

```python
# 先算个简单的
result.clear()
find_deno(fr(1, 57))
sorted(result, key=lambda x:x.denominator)
```

```python
# 验证分子都是1
list(filter(lambda x: x.numerator!=1, result))
```

```python
# 验证分母无重复
len(set(result))==len(result)
```

```python
# 验证求和结果
s = fr(0)
for x in result: s+=x
print(s)
```

```python
# 开工
result.clear()
find_deno(fr(1, 7))
sorted(result, key=lambda x:x.denominator)
```

```python
# 组合成Latex数学格式
s = '$$'
for x in sorted(result, key=lambda x:x.denominator):
    s+='+\\frac{%d}{%d}'%(x.numerator,x.denominator)
s+='$$'
print(s)
```
