```python
def isRegular(a, b, c):
    #判断三角形是否为直角三角形
    blist = [a, b, c]
    blist = sorted(blist)
    return blist[0]**2 + blist[1]**2 == blist[2]**2
```

```python
isRegular(52, 48, 20)
```

```python
def getS(p):
    '''想起我原来的算法了
    上课时用了三层循环，明显只用两层就可以了。
    c应该就是p-a-b，然后做三角形合理性判断，
    两边之和大于第三边，两边之差小于第三边。'''
    return [(a, b, p-a-b) for a in range(1, p/3)
           for b in range(a, p/2)
           if isRegular(a, b, p-a-b) and b<p-a-b and a+b>p-a-b and p-a-b*2<a]
```

```python
getS(120)
```

```python
getS(12)
```
